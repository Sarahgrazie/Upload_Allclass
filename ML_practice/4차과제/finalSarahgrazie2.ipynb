```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# CSV 파일에서 데이터를 Pandas DataFrame으로 불러오기
titanic_data = pd.read_csv('./titanic_train.csv')
titanic_tata1 = pd.read_csv('./titanic_test.csv')

# 'Cabin' 열 삭제
titanic_data = titanic_data.drop(columns='Cabin', axis=1)
titanic_tata1 = titanic_tata1.drop(columns='Cabin', axis=1)

# 'Age' 열의 결측치를 평균값으로 대체
titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True)
titanic_tata1['Age'].fillna(titanic_tata1['Age'].mean(), inplace=True)

# 'Embarked' 열의 결측치를 최빈값으로 대체
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)
titanic_tata1['Embarked'].fillna(titanic_tata1['Embarked'].mode()[0], inplace=True)

# 'Fare' 열의 결측치를 평균값으로 대체
titanic_data['Fare'].fillna(titanic_data['Fare'].mean(), inplace=True)
titanic_tata1['Fare'].fillna(titanic_tata1['Fare'].mean(), inplace=True)

# 범주형 열을 변환
titanic_data.replace({'Sex': {'male': 0, 'female': 1}, 'Embarked': {'S': 0, 'C': 1, 'Q': 2}}, inplace=True)
titanic_tata1.replace({'Sex': {'male': 0, 'female': 1}, 'Embarked': {'S': 0, 'C': 1, 'Q': 2}}, inplace=True)

# 훈련에 불필요한 열 제거
X = titanic_data.drop(columns=['PassengerId', 'Name', 'Ticket', 'Survived'], axis=1)
Y = titanic_data['Survived']

# 훈련 데이터와 테스트 데이터로 나누기
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

# Logistic Regression 모델 생성 및 훈련
model = LogisticRegression()
model.fit(X_train, Y_train)

# 훈련 데이터에 대한 예측
X_train_prediction = model.predict(X_train)

# 훈련 데이터 정확도
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('훈련 데이터 정확도:', training_data_accuracy)

# 테스트 데이터에 대한 예측
X_test_prediction = model.predict(X_test)

# 테스트 데이터 정확도
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('테스트 데이터 정확도:', test_data_accuracy)

# 테스트 데이터에 대한 예측 결과를 DataFrame으로 변환
titanic_tata1['Survived'] = model.predict(titanic_tata1.drop(columns=['PassengerId', 'Name', 'Ticket'], axis=1))
submission = titanic_tata1[['PassengerId', 'Survived']]

# CSV 파일로 저장
submission.to_csv('titanic_submission4.csv', index=False)

# 시각화
sns.set(style="whitegrid", palette="pastel")

# 생존 여부에 따른 Count Plot 생성
ax = sns.countplot(x='Survived', data=titanic_data, palette="Set2")

# 제목과 라벨 추가
ax.set_title('생존 여부 카운트')
ax.set_xlabel('생존 여부')
ax.set_ylabel('인원 수')

# 플롯 표시
plt.show()

# 성별에 따른 Count Plot 생성
ax = sns.countplot(x='Sex', data=titanic_data, palette="Set2")

# 제목과 라벨 추가
ax.set_title('성별에 따른 승객 수')
ax.set_xlabel('성별')
ax.set_ylabel('인원 수')

# 플롯 표시
plt.show()

# 생존 여부 및 성별에 따른 Count Plot 생성
ax = sns.countplot(x='Sex', hue='Survived', data=titanic_data, palette="Set2")

# 제목과 라벨 추가
ax.set_title('성별에 따른 생존 여부 카운트')
ax.set_xlabel('성별')
ax.set_ylabel('인원 수')

# 플롯 표시
plt.show()

# Pclass에 따른 Count Plot 생성
ax = sns.countplot(x='Pclass', data=titanic_data, palette="Set2")

# 제목과 라벨 추가
ax.set_title('Pclass에 따른 승객 수')
ax.set_xlabel('Pclass')
ax.set_ylabel('인원 수')

# 플롯 표시
plt.show()

# Pclass 및 생존 여부에 따른 Count Plot 생성
ax = sns.countplot(x='Pclass', hue='Survived', data=titanic_data, palette="Set2")

# 제목과 라벨 추가
ax.set_title('Pclass에 따른 생존 여부 카운트')
ax.set_xlabel('Pclass')
ax.set_ylabel('인원 수')

# 플롯 표시
plt.show()
```
